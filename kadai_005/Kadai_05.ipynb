```python
import matplotlib.pyplot as plt
import numpy as np

# ===================================================================
# Step 1: üìä ÂÖ®„Å¶„ÅÆ„Ç∞„É©„Éï„ÅÆ„Éá„Éº„Çø„ÇíÊ∫ñÂÇô„Åô„Çã
# ===================================================================

# (1) Êäò„ÇåÁ∑ö„Ç∞„É©„ÉïÁî®„Éá„Éº„Çø
y0_line = [1, 2, -5, 2]

# (2) sinÈñ¢Êï∞Áî®„Éá„Éº„Çø (Â§âÊï∞Âêç„Çí x_sin, y_sin „Å´Â§âÊõ¥)
x_sin = np.linspace(0, 10, 100)
y_sin = 2 + 2 * np.sin(2 * x_sin)

# (3) „Éí„Çπ„Éà„Ç∞„É©„É†Áî®„Éá„Éº„Çø (Â§âÊï∞Âêç„Çí x_hist „Å´Â§âÊõ¥)
x_hist = np.random.normal(15, 5, 2000)

# (4) Êï£Â∏ÉÂõ≥Áî®„Éá„Éº„Çø (Â§âÊï∞Âêç„Çí x_scatter, y_scatter „Å´Â§âÊõ¥)
np.random.seed(3) # ‰π±Êï∞„ÅÆ„Ç∑„Éº„Éâ„ÇíÂõ∫ÂÆö
x_scatter = 4 + np.random.normal(0, 2, 60)
y_scatter = 4 + np.random.normal(0, 2, len(x_scatter))


# ===================================================================
# Step 2: üé® „Çµ„Éñ„Éó„É≠„ÉÉ„ÉàÁî®„ÅÆ„Ç≠„É£„É≥„Éê„ÇπÔºàFigure, AxesÔºâ„ÇíÂÆöÁæ©„Åô„Çã
# ===================================================================

# 2Ë°å x 3Âàó„ÄÅÂêàË®à6„Éû„ÇπÔºà„Çµ„Éñ„Éó„É≠„ÉÉ„ÉàÔºâ„ÅÆ„Ç≠„É£„É≥„Éê„Çπ„ÇíÁî®ÊÑè
# fig: „Ç≠„É£„É≥„Éê„ÇπÂÖ®‰Ωì (Figure)
# axes: ÂêÑ„Éû„Çπ (Axes) „ÅÆÂÖ•„ÇåÁâ©„ÄÇ2Ê¨°ÂÖÉÈÖçÂàó [Ë°å, Âàó] „Å´„Å™„Å£„Å¶„ÅÑ„Åæ„Åô
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 8))

# üìê „Ç∞„É©„Éï„ÅÆ„Çø„Ç§„Éà„É´„ÇÑ„É©„Éô„É´„ÅåÈáç„Å™„Çâ„Å™„ÅÑ„Çà„ÅÜ„Å´Ëá™Âãï„Åß‰ΩôÁôΩ„ÇíË™øÊï¥
fig.tight_layout(pad=3.0)

# ===================================================================
# Step 3: ‚úçÔ∏è ÂêÑ„Çµ„Éñ„Éó„É≠„ÉÉ„ÉàÔºà„Éû„ÇπÔºâ„Å∏„Ç∞„É©„Éï„ÇíÊèèÁîª„Åô„Çã
# ===================================================================

# (1) Êäò„ÇåÁ∑ö„Ç∞„É©„Éï„ÇíÊúÄÂàù„ÅÆ„Éû„Çπ (axs[0] = axes[0, 0]) „Å´ÊèèÁîª
axes[0, 0].plot(y0_line, marker='o') # Ê∫ñÂÇô„Åó„Åü y0_line „Çí‰ΩøÁî®
axes[0, 0].grid(True)


# (2) sinÈñ¢Êï∞„Çí2Áï™ÁõÆ„ÅÆ„Éû„Çπ (axs[1] = axes[0, 1]) „Å´ÊèèÁîª
axes[0, 1].plot(x_sin, y_sin, linewidth=2.0, color='r') # Ê∫ñÂÇô„Åó„Åü x_sin, y_sin „Çí‰ΩøÁî®
axes[0, 1].set_xlim(0, 10)
axes[0, 1].set_ylim(-1, 5)


# (3) „Éí„Çπ„Éà„Ç∞„É©„É†„Çí3Áï™ÁõÆ„ÅÆ„Éû„Çπ (axs[2] = axes[0, 2]) „Å´ÊèèÁîª
axes[0, 2].hist(x_hist, bins=30, color='skyblue', edgecolor='black') # Ê∫ñÂÇô„Åó„Åü x_hist „Çí‰ΩøÁî®

# (4) Êï£Â∏ÉÂõ≥„Çí4Áï™ÁõÆ„ÅÆ„Éû„Çπ (axs[3] = axes[1, 0]) „Å´ÊèèÁîª
axes[1, 0].scatter(x_scatter, y_scatter, color='g', alpha=0.7) # Ê∫ñÂÇô„Åó„Åü x_scatter, y_scatter „Çí‰ΩøÁî®

# -------------------------------------------------------------------
# ‰ΩøÁî®„Åó„Å™„ÅÑÊÆã„Çä„ÅÆ„Éû„ÇπÔºàaxs[4] „Å® axs[5]Ôºâ„ÇíÈùûË°®Á§∫„Å´„Åô„Çã
# -------------------------------------------------------------------
#axes[4].set_visible(False)
#axes[5].set_visible(False)


# ===================================================================
# Step 4: üñ•Ô∏è „Ç∞„É©„Éï„ÇíË°®Á§∫„Åô„Çã
# ===================================================================

plt.show()
```
