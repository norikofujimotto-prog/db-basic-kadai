```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib

# データの読み込み
df = pd.read_csv("sample_pandas_6.csv")

# データの確認 (最初の5行)
print(df.head())
```

```python
# '商品番号'の出現頻度をカウント
# 結果はシリーズ (Series) として返され、インデックスが商品番号、値が頻度になる
frequency = df['商品番号'].value_counts()

# カウント結果の確認
print("\n--- 商品番号の出現頻度 ---\n", frequency)
```

```python
# 棒グラフの描画
plt.figure(figsize=(10, 6)) # グラフのサイズを設定

# 棒グラフを作成
# x軸: 商品番号 (インデックス), y軸: 頻度 (値)
frequency.plot(kind='bar', color='skyblue')

# グラフのタイトルとラベルを設定
plt.title('商品番号ごとの注文出現頻度', fontsize=16)
plt.xlabel('商品番号', fontsize=12)
plt.ylabel('出現回数（頻度）', fontsize=12)

# x軸の目盛りを45度回転させて見やすくする
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7) # 横軸に補助線を追加
plt.tight_layout() # ラベルがはみ出さないように調整

# グラフを表示
plt.show()
```

```python
import pandas as pd

# データの読み込み
df = pd.read_csv("sample_pandas_6.csv")

# 1. 商品番号 (基準列) でデータをグループ化する
# 2. 注文数 (サブ列) を選択する
# 3. describe() メソッドを使って基本統計量 (集計) を計算する

summary_stats = df.groupby('商品番号')['注文数'].describe()

# 結果を表示
print(summary_stats)
```
